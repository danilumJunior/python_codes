```python
'''10 – A partir da leitura de um número indeterminado de
valores não nulos, determinar e exibir quantos desses
valores são positivos e quantos são negativos. Determinar
também a soma dos positivos e a soma dos negativos. O
último valor a ser lido é zero (0 é o flag).'''

positivos = 0
negativos = 0
soma_positivos = 0
soma_negativos = 0


while True:
    numero = float(input("Digite um número (0 para encerrar): "))

    
    if numero == 0:
        break

   
    if numero > 0:
        positivos += 1
        soma_positivos += numero
    else:
        negativos += 1
        soma_negativos += numero


print(f"Quantidade de números positivos: {positivos}")
print(f"Quantidade de números negativos: {negativos}")
print(f"Soma dos números positivos: {soma_positivos}")
print(f"Soma dos números negativos: {soma_negativos}")
```

```python
'''11 – A partir da leitura de um numero indeterminado de notas,
calcular e exibir a média aritmética dessas notas. O flag
para indicar o fim das notas é –1.'''
'''12 – O mesmo exercício anterior fornecendo adicionalmente a
maior e a menor nota lida.'''

soma_notas = 0
quantidade_notas = 0
maior_nota = float('-inf')  
menor_nota = float('inf')  


while True:
    nota = float(input("Digite a nota (-1 para encerrar): "))

    
    if nota == -1:
        break

    
    soma_notas += nota
    quantidade_notas += 1
    if nota > maior_nota:
        maior_nota = nota
    if nota < menor_nota:
        menor_nota = nota


if quantidade_notas > 0:
    media = soma_notas / quantidade_notas
    print(f"A média das notas é: {media:.2f}")
    print(f"A maior nota é: {maior_nota}")
    print(f"A menor nota é: {menor_nota}")
else:
    print("Nenhuma nota foi inserida.")
```

```python
'''Escreva um programa que calcule o fatorial de um número dado.'''

def calcular_fatorial(numero):
    if numero < 0:
        return "Fatorial não definido para números negativos"
    elif numero == 0 or numero == 1:
        return 1
    else:
        resultado = 1
        for i in range(2, numero + 1):
            resultado *= i
        return resultado


numero = int(input("Digite um número para calcular o fatorial: "))


resultado_fatorial = calcular_fatorial(numero)
print(f"O fatorial de {numero} é: {resultado_fatorial}")


```

```python
'''Implemente um algoritmo de ordenação para ordenar uma lista de valores.'''

def ordenar_lista(lista): # Essa função implementa o algoritmo de ordenação por inserção.
    for i in range(1, len(lista)): #  Este é um loop que percorre a lista a partir do segundo elemento (i = 1) até o último (len(lista) - 1). Isso permite que o algoritmo de ordenação por inserção seja aplicado.
        chave = lista[i] # Aqui, está selecionando o elemento atual que será inserido na parte ordenada da lista
        j = i - 1 # Inicializamos j como o índice anterior ao elemento atual.
        while j >= 0 and chave < lista[j]: #Este é um loop while que verifica se o elemento chave é menor que o elemento à esquerda (lista[j]) na parte ordenada da lista. Ele move os elementos maiores para a direita para abrir espaço para o elemento chave.
            lista[j + 1] = lista[j]
            j -= 1
        lista[j + 1] = chave #Finalmente, colocamos o elemento chave na posição correta na parte ordenada da lista.

# Pedir ao usuário para digitar os valores separados por vírgula
entrada = input("Digite os valores separados por vírgula: ")
valores = [int(valor) for valor in entrada.split(",")]

# Chamar a função de ordenação
ordenar_lista(valores)
print("Lista ordenada:", valores)
```

```python
''' Crie um programa que identifique se uma palavra é um palíndromo.'''

def eh_palindromo(palavra):
    # Remover espaços em branco e tornar tudo minúsculo
    palavra = palavra.replace(" ", "").lower()
    
    # Verificar se a palavra é igual à sua inversa
    return palavra == palavra[::-1]

# Solicitar ao usuário para digitar uma palavra
palavra = input("Digite uma palavra: ")

# Verificar se a palavra é um palíndromo e exibir o resultado
if eh_palindromo(palavra):
    print(f'A palavra "{palavra}" é um palíndromo.')
else:
    print(f'A palavra "{palavra}" não é um palíndromo.')
```

```python
'''Desenvolva um algoritmo para verificar se dois conjuntos possuem os mesmos elementos.'''

def mesmos_elementos(conjunto1, conjunto2):
    return conjunto1 == conjunto2

# Exemplo de uso:
conjunto1 = {1, 2, 3, 4, 5}
conjunto2 = {3, 2, 1, 5, 4}
if mesmos_elementos(conjunto1, conjunto2):
    print("Os conjuntos possuem os mesmos elementos.")
else:
    print("Os conjuntos não possuem os mesmos elementos.")
```

```python
'''Dada uma sequência de números de 1 a N, encontre o número que está faltando.'''


def encontrar_numero_faltando(sequencia):
    # Calcular a soma esperada da sequência completa de 1 a N
    n = len(sequencia) + 1  # Adiciona 1 para incluir o número faltando
    soma_esperada = n * (n + 1) // 2

    # Calcular a soma real dos números dados na sequência
    soma_real = sum(sequencia)

    # Encontrar o número faltando
    numero_faltando = soma_esperada - soma_real
    return numero_faltando

# Solicitar ao usuário para digitar os valores da sequência separados por vírgula
entrada = input("Digite os valores da sequência de 1 a N separados por vírgula: ")
sequencia = [int(numero) for numero in entrada.split(",")]

# Encontrar o número faltando na sequência e exibir o resultado
numero_faltando = encontrar_numero_faltando(sequencia)
print(f"O número faltando na sequência é: {numero_faltando}")
```

```python
''' Implemente um algoritmo que converta um número decimal para seu equivalente em binário.'''

def decimal_para_binario(decimal):
    if decimal == 0:
        return '0'  # Caso especial para o número 0

    binario = ''
    while decimal > 0:
        resto = decimal % 2
        binario = str(resto) + binario
        decimal //= 2
    return binario

# Exemplo de uso:
numero_decimal = int(input("Digite um número decimal: "))
numero_binario = decimal_para_binario(numero_decimal)
print(f'O número {numero_decimal} em binário é: {numero_binario}')
```
